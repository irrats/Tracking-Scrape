```python
# Import packages
import requests
from bs4 import BeautifulSoup

# Specify url
url = 'https://trackapkg.com'

# Package the request, send the request and catch the response: r
r = requests.get(url)

# Extracts the response as html: html_doc
html_doc = r.text

# create a BeautifulSoup object from the HTML: soup
soup = BeautifulSoup(html_doc)

# Print the title of Guido's webpage
print(soup.title)

# Find all 'a' tags (which define hyperlinks): a_tags
a_tags = soup.find_all('a')

# Print the URLs to the shell
for link in a_tags:
    print(link.get('href'))
```

```python
        ####### Part 1: Loop to Extract titles from each Company

import pandas as pd

import requests
from bs4 import BeautifulSoup

# Specify url
url = 'https://trackapkg.com/aramex-tracking-number' #'https://trackapkg.com/a.duiepyle-tracking-number' 

# Package the request, send the request and catch the response: r
r = requests.get(url)

# Extracts the response as html: html_doc
html_doc = r.text

# create a BeautifulSoup object from the HTML: soup
soup = BeautifulSoup(html_doc)

# Print the title (only text with no extra spaces/ tags etc.)
title = soup.title
title_text = title.get_text().strip()
print(title_text)


file = pd.read_csv('test_track.csv', index_col=0, names=["company"])
df = pd.DataFrame(file)
comname = []

for row in df['company']:
    url = row
    # Package the request, send the request and catch the response: r
    r = requests.get(url)

    # Extracts the response as html: html_doc
    html_doc = r.text

    # create a BeautifulSoup object from the HTML: soup
    soup = BeautifulSoup(html_doc)

    # Print the title (only text with no extra spaces/ tags etc.)
    title = soup.title
    title_text = title.get_text().strip()
    comname.append(title_text)
    
df['title'] = comname
print(df)
```

```python
             ########### Part 2: Extract titles and Onclick URLs to Dataframe (if sees submit button passes an exception)

import pandas as pd

import requests
from bs4 import BeautifulSoup
import re

file = pd.read_csv('test_track.csv', index_col=0, names=["company"])
df = pd.DataFrame(file)
comname = []
comurl = []

for row in df['company']:
    url = row
    # Package the request, send the request and catch the response: r
    r = requests.get(url)

    # Extracts the response as html: html_doc
    html_doc = r.text

    # create a BeautifulSoup object from the HTML: soup
    soup = BeautifulSoup(html_doc)

    # Print the title (only text with no extra spaces/ tags etc.)
    title = soup.title
    title_text = title.get_text().strip()
    comname.append(title_text)
    
df['title'] = comname

for row in df['company']:
    try:
        url = row
        # Package the request, send the request and catch the response: r
        r = requests.get(url)

        # Extracts the response as html: html_doc
        html_doc = r.text

        # create a BeautifulSoup object from the HTML: soup
        soup = BeautifulSoup(html_doc)

        # Create a pattern of what text we're searching under the tag http:
        pattern = re.compile(r"['\"](http(?:s?)://[^'\"]+)")

        #Using Css locator we get to desired tag, choose onclick and print patterned text
        data = soup.select_one("span.input-group-btn>input[onclick^='window.open']")["onclick"]
        com_url = re.search(pattern, data).group(1)

        comurl.append(com_url)
    except:
        comurl.append('oops')
df['adrs'] = comurl
print(df)
```
